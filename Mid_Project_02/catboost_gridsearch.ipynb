## 스플릿

```python
from sklearn.model_selection import train_test_split
```

```python
# train_test_split?
```

```python
X_lunch_train, X_lunch_test, y_lunch_train, y_lunch_test = train_test_split(
    train_lunch.drop(columns=label_lunch), train_lunch[label_lunch], test_size=0.1, random_state=42)
```

```python
X_dinner_train, X_dinner_test, y_dinner_train, y_dinner_test = train_test_split(
    train_dinner.drop(columns=label_dinner), train_dinner[label_dinner], test_size=0.1, random_state=42)
```

```python
X_dinner_train.shape, X_dinner_test.shape, y_dinner_train.shape, y_dinner_test.shape
```

## Catboost + GridSearch + Cross Validation

### 파라미터 설정

```python
# set_n_estimators = list(range(250, 410, 20))
set_n_estimators = list(range(250, 610, 20))
# set_n_estimators = [310]
print(set_n_estimators)
```

```python
# set_learning_rate = [lr / 100 for lr in list(range(0, 10, 1))]
# set_learning_rate = [lr / 1000 for lr in list(range(85, 95, 1))]
set_learning_rate = [lr / 100 for lr in list(range(10, 100, 1))]
# set_learning_rate = []
print(set_learning_rate)
```

```python
param_grid = {
    'n_estimators': set_n_estimators,
    'learning_rate': set_learning_rate
}

print(param_grid)
```

```python
import catboost
# from sklearn.model_selection import cross_validate
```

### lunch model

```python
lunch_model = catboost.CatBoostRegressor(
    eval_metric='MAE',
    random_state=42,
    logging_level='Silent',
    early_stopping_rounds=200
)
```

```python
lunch_result = lunch_model.grid_search(
    param_grid=param_grid,
    X=X_lunch_train,
    y=y_lunch_train,
    cv=10
)
lunch_result
```

```python
lunch_model.get_all_params()
```

```python
lunch_model.feature_names_
```

```python
lunch_model.feature_importances_
```

```python
import koreanize_matplotlib
```

```python
pd.DataFrame(
    lunch_model.feature_importances_, index=lunch_model.feature_names_
).sort_values(0).plot.barh()
```

```python
df_lunch_result = pd.DataFrame(lunch_result)
df_lunch_result
```

```python
pd.DataFrame({
    'train-MAE-mean': df_lunch_result.loc['train-MAE-mean', 'cv_results'],
    'test-MAE-mean': df_lunch_result.loc['test-MAE-mean', 'cv_results']
}).plot();
```

### dinner model

```python
dinner_model = catboost.CatBoostRegressor(
    eval_metric='MAE',
    random_state=42,
    logging_level='Silent',
    early_stopping_rounds=200
)
```

```python
dinner_result = dinner_model.grid_search(
    param_grid=param_grid,
    X=X_dinner_train,
    y=y_dinner_train,
    cv=10
)
dinner_result
```

```python
dinner_model.get_all_params()
```

```python
df_dinner_result = pd.DataFrame(dinner_result)
df_dinner_result
```

```python
pd.DataFrame({
    'train-MAE-mean': df_dinner_result.loc['train-MAE-mean', 'cv_results'],
    'test-MAE-mean': df_dinner_result.loc['test-MAE-mean', 'cv_results']
}).plot();
```
