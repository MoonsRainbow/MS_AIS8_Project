## 중식메뉴 카운트

```python
lunch = train['중식메뉴'].str.split().tolist()
```

```python
lunch[0]
```

```python
for meal in lunch:
    del_list = []
    for i, menu in enumerate(meal):
        if str(menu).startswith('(') and str(menu).endswith(')'):
            del_list.append(i)
    for i in del_list[::-1]:
        del meal[i]
```

```python
lunch[0]
```

```python
for meal in lunch:
    del_list = []
    for i, menu in enumerate(meal):
        if menu == '(':
            del_list.append(i)
    for i in del_list[::-1]:
        del meal[i]
```

```python
lunch[0]
```

```python
for meal in lunch:
    add_list = []
    del_list = []
    for i, menu in enumerate(meal):
        if '/' in menu and '밥' not in menu:
            add_list = menu.replace('/', ' ').split()
            del_list.append(i)
    for i in del_list[::-1]:
        del meal[i]
    meal += add_list
```

```python
for meal in lunch:
    del_list = []
    for i, menu in enumerate(meal):
        if menu.startswith('*'):
            del_list.append(i)
    for i in del_list[::-1]:
        del meal[i]
```

```python
menu_count = []
for meal in lunch:
    menu_count.append(len(meal))
print(menu_count)
```

```python
dinner = train['석식메뉴'].str.split().tolist()
```

```python
for meal in dinner:
    del_list = []
    for i, menu in enumerate(meal):
        if str(menu).startswith('(') and str(menu).endswith(')'):
            del_list.append(i)
    for i in del_list[::-1]:
        del meal[i]
```

```python
train['일자'] = pd.to_datetime(train['일자'])
```

```python
train['year'] = train['일자'].dt.year
```

```python
train['month'] = train['일자'].dt.month
```

```python
train['day'] = train['일자'].dt.day
```

```python
train['quarter'] = train['일자'].dt.quarter
```

```python
train['year_month'] = train['일자'].astype('str').str[:7]
```

```python
train.groupby('year_month').agg({
    '중식계': 'mean',
    '중식메뉴_카운트': 'mean'
}).plot(
    title='연월 별 중식 식사 인원 & 메뉴 개수 그래프', 
    secondary_y='중식메뉴_카운트',
    xlabel='연월',
    ylabel='식사 인원 (왼쪽, 명), 메뉴 개수(오른쪽, 개)'
);
```

```python
train.groupby('year_month').agg({
    '석식계': 'mean',
    '석식메뉴_카운트': 'mean'
}).plot(
    title='연월 별 석식 식사 인원 & 메뉴 개수 그래프', 
    secondary_y='석식메뉴_카운트',
    xlabel='연월',
    ylabel='식수(왼쪽, 명), 메뉴 개수(오른쪽, 개)'

);
```
