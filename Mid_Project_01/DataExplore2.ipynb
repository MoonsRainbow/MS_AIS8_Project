```python
import numpy as np
import pandas as pd
import seaborn as sns
import koreanize_matplotlib
import matplotlib.pyplot as plt
```

```python
plt.ion()
plt.style.use('bmh')
```

```python
df_origin = pd.read_csv('./data/all_201703-202209.csv')
```

```python
df_origin = df_origin.reset_index(drop=True)
```

```python
df_pre = df_origin.copy(deep=True)
```

```python
df_pre = df_pre.drop(columns=['회사명', '회사코드', '보고서명', '보고서코드', 'factor_id', '단위'])
```

```python
df_pre = df_pre.set_index('항목')
```

```python
df_pre = df_pre.T
```

```python
df_pre = df_pre.astype('float')
```
