- Ladder:

    - 1인당 GDP

    - 2020년, 2021년은 'Explained by: Log GDP per capita'

- Social support:
    - 해당 국가의 사회적 지원에 대한 점수 지표

<br><br>

- Healthy life expectancy:
    - 기대 수명에 대한 점수 지표

- Freedom to make life choices:

<br><br>

- Generosity:
    - 관대함에 대한 점수 지표
    - 2019년의 'Generosity'
    - 2020년, 2021년은 'Explained by: Generosity'
<br><br>

- Perceptions of corruption:
    - 부패에 대한 인식 점수 지표
    - 2019년의 'Perceptions of corruption'
    - 2020년, 2021년은 'Explained by: Perceptions of corruption'

```python
integrated_columns = [
    'Year',
    'Regional indicator',
    'Country name',
    'Ladder',
    'Happy score',
    'GDP per capita',
    'Social support',
    'Healthy life expectancy',
    'Freedom to make life choices',
    'Generosity',
    'Perceptions of corruption'
]
```

### 2019년도 데이터 전처리
<br>

1. 데이터 프레임 생성 및 존재하지 않는 컬럼 생성
1. 통합 컬럼으로 컬럼 생성, 데이터 추가
1. 컬럼명이 변경된 데이터 수동 추가
<br><br>

- Regional indicator, Country Name은 2020년, 2021년 전처리 후 처리

```python
# 데이터 프레임 생성 및 존재하지 않는 컬럼 생성
df_happy_19 = pd.DataFrame(columns=[
    'Year', 
    'RI', 
    'CN', 
    'Ladder', 
    'Happy score'
])
df_happy_19
```

```python
# 통합 컬럼으로 컬럼 생성, 데이터 추가
df_happy_19 = pd.concat([df_happy_19, happy_19[integrated_columns[5:]]], axis=1)
df_happy_19.head()
```

```python
# 컬럼명이 변경된 데이터 수동 추가
df_happy_19[['Ladder', 'Happy score', 'CN']] = happy_19[['Overall rank', 'Score', 'Country or region']]
df_happy_19['Year'] = 2019
df_happy_19.head()
```

```python
# 데이터 확인
display(happy_19.head())
display(df_happy_19.head())
df_happy_19.info()
```

### 2020년 데이터 전처리
<br>

1. 데이터 프레임 생성 및 컬럼 생성
1. 통합 컬럼으로 데이터 추가

```python
# 데이터 프레임 생성 및 컬럼 생성
df_happy_20 = pd.DataFrame(columns=[
    'Year',
    'Ladder',
    'Happy score',
    'GDP per capita',
    'Social support',
    'Healthy life expectancy',
    'Freedom to make life choices',
    'Generosity',
    'Perceptions of corruption'
])
df_happy_20
```

```python
# 통합 컬럼으로 데이터 추가
df_happy_20['Ladder'] = happy_20['Ladder score'].rank(method='dense', ascending=False).astype(int)
df_happy_20[['RI', 'CN', 'Happy score']] = happy_20[['Regional indicator', 'Country name', 'Ladder score']]
df_happy_20[integrated_columns[5:]] = happy_20[happy_20.columns[-7:-1]]
df_happy_20['Year'] = 2020
df_happy_20.head()
```

```python
# 실수형 데이터, 소수점 3자리로 고정
df_happy_20.iloc[:, -7:] = round(df_happy_20.iloc[:, -7:], 3)
df_happy_20.head()
```

```python
# 데이터 확인
display(happy_20.head())
display(df_happy_20.head())
df_happy_20.info()
```

### 2021년 데이터 전처리
<br>

1. 데이터 프레임 생성 및 컬럼 생성
1. 통합 컬럼으로 데이터 추가

```python
# 데이터 프레임 생성 및 컬럼 생성
df_happy_21 = pd.DataFrame(columns=[
    'Year',
    'Ladder',
    'Happy score',
    'GDP per capita',
    'Social support',
    'Healthy life expectancy',
    'Freedom to make life choices',
    'Generosity',
    'Perceptions of corruption'
])
df_happy_21
```

```python
# 통합 컬럼으로 데이터 추가
df_happy_21['Ladder'] = happy_21['Ladder score'].rank(method='dense', ascending=False).astype(int)
df_happy_21[['RI', 'CN', 'Happy score']] = happy_21[['Regional indicator', 'Country name', 'Ladder score']]
df_happy_21[integrated_columns[5:]] = happy_21[happy_21.columns[-7:-1]]
df_happy_21['Year'] = 2021
df_happy_21.head()
```

```python
# 데이터 확인
display(happy_21.head())
display(df_happy_21.head())
df_happy_21.info()
```

### Regional indicator, Country name 전처리
<br>

1. 연도별 Country name 유니크 확인
1. 각 연도별 Regional indicator, Country name만 추출
1. 가장 데이터가 많은 데이터프레임을 기준으로 머지
1. 전처리가 필요한 데이터 확인
1. 전처리가 필요 없는 데이터 확인
1. 통합 데이터 컬럼 생성
1. 전처리 확인

```python
# 각 연도별 Regional indicator, Country name만 추출
df_19_rc = df_happy_19[['RI', 'CN']].copy(deep=True)
df_20_rc = df_happy_20[['RI', 'CN']].copy(deep=True)
df_21_rc = df_happy_21[['RI', 'CN']].copy(deep=True)

df_19_rc.columns = ['RI-19', 'CN-19']
df_20_rc.columns = ['RI-20', 'CN-20']
df_21_rc.columns = ['RI-21', 'CN-21']

display(df_19_rc)
display(df_20_rc)
display(df_21_rc)
```

```python
# 가장 데이터가 많은 데이터프레임을 기준으로 머지
df_rc = df_19_rc.merge(df_20_rc, how='outer', left_on='CN-19', right_on='CN-20')
df_rc = df_rc.merge(df_21_rc, how='outer', left_on='CN-19', right_on='CN-21')
display(df_rc.shape)
display(df_rc.head())
display(df_rc.tail())
```

```python
# 전처리가 필요한 데이터 확인
df_rc['Need pre-process'] = df_rc['CN-19'].isnull() | df_rc['CN-20'].isnull() | df_rc['CN-21'].isnull()
display(df_rc.loc[df_rc['Need pre-process']].shape)
display(df_rc.loc[df_rc['Need pre-process']].head())
display(df_rc.loc[df_rc['Need pre-process']].tail())
```

```python
# 전처리가 필요 없는 데이터 확인
df_rc.loc[~df_rc['Need pre-process']]
display(df_rc.loc[~df_rc['Need pre-process']].shape)
display(df_rc.loc[~df_rc['Need pre-process']].head())
display(df_rc.loc[~df_rc['Need pre-process']].tail())
```

```python
# 통합 데이터 컬럼 생성
df_rc[['Regional indicator', 'Country name']] = df_rc.loc[~df_rc['Need pre-process'], ['RI-20', 'CN-20']]

# Taiwan 컬럼 통합
df_rc.loc[[24, 156, 162], ['Regional indicator', 'Country name']] = ['East Asia', 'Taiwan']

# North Cyprus 컬럼 통합
df_rc.loc[[63, 158, 163], ['Regional indicator', 'Country name']] = ['Western Europe', 'North Cyprus']

# Hong Kong 컬럼 통합
df_rc.loc[[75, 159, 164], ['Regional indicator', 'Country name']] = ['East Asia', 'Hong Kong']

# North Macedonia 컬럼 통합
df_rc.loc[[83, 161], ['Regional indicator', 'Country name']] = ['Central and Eastern Europe', 'North Macedonia']

# 2019, 2020, 2021년 csv 중 한 곳이라도 없는 국가 처리
df_rc.loc[
    df_rc['Need pre-process'] & df_rc['Regional indicator'].isnull(), 
    ['Regional indicator', 'Country name']
] = ['Delete', 'Delete']

# 전처리 완료
df_rc
```
